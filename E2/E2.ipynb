**A CONTINUACIÓN SE PROCEDE A LA FILTRACIÓN DE DATOS:**

 Se eliminan las columnas que no son necesarias: 'Codigo region' - 'Comuna' - 'Codigo comuna' - 'Población'

```python
casos.drop(['Codigo region','Comuna','Codigo comuna','Poblacion'], axis=1,inplace=True)
```

```python
casos
```

**DESARROLLO:**
1. Casos Confirmados por Región, es decir la suma por comuna, en el tiempo.

Definimos una varible 'regiones' para almacenar los datos de Región, con el fin de mostrar solamente una vez la Región.

```python
regiones=casos.Region.unique()
```

```python
regiones
```

Creamos la variable 'agrupados' con el propósito de almacenar los datos a través de la función groupby que nos permite agrupar los datos y realizar operaciones con estos.

```python
agrupados=casos.groupby(by="Region").sum()    
```

```python
x=np.arange(0,agrupados.T.shape[0])
```

```python
sns.set_theme()
sns.set_style('dark')
```

**GRÁFICO N°1**

```python
agrupados.T.plot(figsize=(20,10),rot=90,xticks=x,xlabel='SEMANAS EPIDEMIOLÓGICAS',ylabel='CANTIDAD DE CASOS',title='CASOS CONFIRMADOS POR REGIÓN')
```

2. Casos confirmados, pero el acumulado.

**Se utiliza la función *cumsum* para calcular el valor acumulado de los casos confirmados por semana epidemiológica**

```python
acumulados=agrupados.cumsum(axis=1)
```

```python
x=np.arange(0,acumulados.T.shape[0])
```

```python
acumulados
```

**GRÁFICO N°2**

```python
acumulados.T.plot(figsize=(20,10),xticks=x, xlabel='SEMANAS EPIDEMIOLÓGICAS', ylabel="CANTIDAD DE CASOS" , 
         title = "CASOS ACUMULADOS CONFIRMADOS ", rot=90)
```
